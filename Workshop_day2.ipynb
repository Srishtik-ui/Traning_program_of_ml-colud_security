```python
import random
import math

heights = [random.randint(150,190) for i in range(20)]
print("Sample height (cm):", heights)
#snd
#mean nd sd
mean = sum(heights)/len(heights)
variance = sum([(i-mean)**2 for i in heights])/len(heights)
sd = math.sqrt(variance)
print("Standard deviation:", sd)
print("Mean height (cm):", mean)
```

```python
import random
import math

random.seed(43)
heights = [random.randint(150,190) for i in range(20)]
print("Original sample heights (cm):", heights)


heights.insert(0, 1000)
print("Sample heights with outliers (cm):", heights)

mean = sum(heights)/len(heights)
variance = sum([(i-mean)**2 for i in heights])/len(heights)
sd = math.sqrt(variance)
print("\nStandard deviation with outliers:", sd)
print("Mean height with outliers (cm):", mean)


Z_score_first_element = (heights[0]-mean)/sd
print("\nFirst element height:", heights[0])
print("Z_score for first element:", Z_score_first_element)

z_scores = [(h - mean) / sd for h in heights]
print("\nZ-scores for all heights:", z_scores)

outlier_value_1000 = 1000
```
