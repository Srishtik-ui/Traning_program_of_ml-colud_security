```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])
print("\na",type(a))
print(a)
print("\nb",type(b))
print(b)

#matrix as matrix object

c = np.matrix([[1,2],[3,4]])
d = np.matrix([[5,6],[7,8]])
print("\nc",type(c))
print(c)
print("\nd",type(d))
print(d)
print("\n* operation on two ndarray object (ElementWise)")
print(a * b)
print("\n operation on two matrix object (same as np.dot())")
print(c * d)
```

```python
#Rank of matrix
a = np.arange(1,10)
a.shape = (3,3)
print("a = ")
print(a)
print("rank of a = ")
print(np.linalg.matrix_rank(a))
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])


det = A[0,0]*A[1,1] - A[0,1]*A[1,0]

if det == 0:
    print("Inverse does not exist")
else:

    cofactor = np.array([[ A[1,1], -A[0,1]],
                          [-A[1,0],  A[0,0]]])

    adjoint = cofactor.T
    inverse = (1/det) * adjoint

    print("Inverse Matrix:")
    print(inverse)
```

```python
#pseudo Inverse
a = np.array([[2,8],
 [1,4]])
print("a = ")
print(a)
det = np.linalg.det(a)
print("Determinant:",det)
pinv = np.linalg.pinv(a)
print("\nPseudo Inverse of a = ")
print(pinv)
```

```python
# Flattening a Matrix
a = np.arange(1,10)
a.shape = (3,3)
print("a = ")
print(a)
print("\nAfter flattening: ")
print(a.flatten())
print("\nAfter reshaping:")
print(a.reshape(-1))
```
