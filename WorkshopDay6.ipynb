```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.randint(1, 101, size = 200)
print(data)
```

```python
#histrogram
plt.hist(data, bins = 30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
categories = ['A','B','C']
category_count = [300,400,500]
plt.bar(categories, category_count)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Plot')
plt.show()
```

```python
from sklearn import datasets
import matplotlib.pyplot as plt
iris = datasets.load_iris()
sepal_length = iris.data [:, 0]

plt.boxplot(sepal_length, labels = ['Sepal Length'])
plt.ylabel('cm')
plt.title('Box Plot of sepal length')
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(sepal_length, shade = True)
plt.xlabel('cm')
plt.ylabel('Density')
plt.title('Density Plot of sepal length')
plt.show()
```

```python
plt.violinplot(data, showmedians = True)
plt.ylabel('Value')
plt.title('Violin Plot of Random Data')
plt.show()
```

```python
plt.violinplot(iris.data, showmedians = True)
plt.ylabel('Value')
plt.title('Violin Plot of Random Data')
plt.show()
```

```python
iris = datasets.load_iris()
species = iris.target_names
species_counts = np.bincount(iris.target)
plt.figure(figsize = (8,8))
plt.pie(species_counts, labels = species, autopct = '%1.1f%%', startangle = 90)
plt.title('Pie Chart of Iris Species')
plt.show()
```

```python
iris = datasets.load_iris()
species = iris.target_names
species_counts = np.bincount(iris.target)

#Donut charts
plt.figure(figsize = (8,8))
plt.pie(species_counts, labels = species, autopct = '%1.1f%%', startangle = 90, wedgeprops = {'width': 0.5})
plt.title('Species Distribution in Iris Dataset(Donut Chart)')
plt.show()
```

```python
color = ['#ff9999', '#ffc000','#8fd9b6']
explode = (0,0.1,0.1)

plt.pie(category_count, explode= explode, labels = categories, colors=color, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.axis('equal')
plt.title('fruit Distribution')
plt.show()
```

```python
iris = datasets.load_iris()

df1 = pd.DataFrame(iris.data, columns = iris.feature_names)
df1['species'] = iris.target_names[iris.target]

plt.figure(figsize= (8,6))
for species in iris.target_names:
   sns.kdeplot(data = df1[df1['species'] == species], x = 'sepal length (cm)', label = species, fill = True)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.title('Distribution of Sepal Length by Species')
plt.legend()
plt.show()
```

```python
X = ['Section A','Section B','Section C','Section D']
Ygirls = [10,20,30,40]
Zboys = [20,30,20,30]
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'Girls',color = 'r')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Boys')
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Student")
plt.title("Number of Student in each group")
plt.legend()
plt.show()
```

```python
data = {
    'city' : ['City 1','City 2','City 3'],
    'population' : [25,28,30],
    'Avg_Income' : [40000,50000,60000],
    'Pollution_Index' : [55,66,80]
}
df = pd.DataFrame(data)
df

plt.figure(figsize = (3,3))
plt.scatter(df['Avg_Income'],df['Pollution_Index'],s = df['population']* 0.5, alpha = 0.5, c = df['population'])
plt.colorbar(label = 'Population')
plt.xlabel('Average Income (In $1000)')
plt.ylabel('Pollution Index')
plt.title('City Bubble PLot : Income vs Pollution')


for i in range(len(df)):
  plt.annotate(df['city'][i], (df['Avg_Income'][i],df['Pollution_Index'][i]),fontsize= 2,ha ='right')

plt.show()
```
