```python
def print_names(*names):
  for name in names:
    print(name)

print_names("srishti", "kashish", "shubhra", "harshita")
```

The `*` operator can be used in a few ways for unpacking:

1.  **Unpacking in assignments:** To assign parts of an iterable to different variables.
2.  **Function arguments (`*args`):** To allow a function to accept an arbitrary number of positional arguments.

```python
list1 = [1,2,3]
list2 = [4,5,6]

#list1.append(list2)
#list1.extend(list2)
print(list1+list2)
#print(list1,list2)
```

```python
dict1 = {"a":1,"b":2}
dict2 = {"c":3,"d":4}

combined_dict = {**dict1,**dict2}
print(f"combined dictionary: {combined_dict}\n")
```
